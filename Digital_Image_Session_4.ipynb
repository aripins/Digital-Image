```python
import cv2
from matplotlib import pyplot as plt

# Baca gambar grayscale
img = cv2.imread('ningning.jpeg', 0)

# Histogram Equalization
equ = cv2.equalizeHist(img)

# Buat histogram dari gambar asli dan hasil equalization
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([equ], [0], None, [256], [0, 256])

# Tampilkan gambar asli dan hasil equalization
plt.figure(figsize=(10, 5))

# Gambar asli
plt.subplot(221), plt.imshow(img, cmap='gray'), plt.title('Gambar Asli')
plt.subplot(222), plt.imshow(equ, cmap='gray'), plt.title('Hasil Equalization')

# Histogram gambar asli
plt.subplot(223), plt.plot(hist_original, color='black'), plt.title('Histogram Gambar Asli')
plt.xlim([0, 256])

# Histogram gambar hasil equalization
plt.subplot(224), plt.plot(hist_equalized, color='black'), plt.title('Histogram Hasil Equalization')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Baca gambar
img = cv2.imread('ningning.jpeg')

# Tambahkan Gaussian noise pada gambar (secara manual, menggunakan normal distribution)
mean = 0
sigma = 25
gaussian_noise = np.random.normal(mean, sigma, img.shape).astype(np.uint8)
noisy_img = cv2.add(img, gaussian_noise)  # Menambahkan noise ke gambar asli

# Terapkan Gaussian Blur untuk Noise Reduction
blurred = cv2.GaussianBlur(noisy_img, (5, 5), 0)

# Konversi gambar dari BGR ke RGB untuk Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
noisy_img_rgb = cv2.cvtColor(noisy_img, cv2.COLOR_BGR2RGB)
blurred_rgb = cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB)

# Konversi gambar dari BGR ke grayscale untuk histogram
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
noisy_img_gray = cv2.cvtColor(noisy_img, cv2.COLOR_BGR2GRAY)
blurred_gray = cv2.cvtColor(blurred, cv2.COLOR_BGR2GRAY)

# Buat histogram dari gambar asli, gambar dengan noise, dan hasil Gaussian Blur
hist_img = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
hist_noisy_img = cv2.calcHist([noisy_img_gray], [0], None, [256], [0, 256])
hist_blurred_img = cv2.calcHist([blurred_gray], [0], None, [256], [0, 256])

# Tampilkan gambar asli, gambar dengan noise, dan hasil Gaussian Blur serta histogram masing-masing
plt.figure(figsize=(14, 8))

# Gambar asli
plt.subplot(231), plt.imshow(img_rgb), plt.title('Gambar Asli')
plt.axis('off')
plt.subplot(234), plt.plot(hist_img, color='black'), plt.title('Histogram Gambar Asli')
plt.xlim([0, 256])

# Gambar dengan noise
plt.subplot(232), plt.imshow(noisy_img_rgb), plt.title('Gambar dengan Noise')
plt.axis('off')
plt.subplot(235), plt.plot(hist_noisy_img, color='black'), plt.title('Histogram Gambar dengan Noise')
plt.xlim([0, 256])

# Gambar hasil Gaussian Blur
plt.subplot(233), plt.imshow(blurred_rgb), plt.title('Hasil Gaussian Blur')
plt.axis('off')
plt.subplot(236), plt.plot(hist_blurred_img, color='black'), plt.title('Histogram Hasil Gaussian Blur')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

# Baca gambar grayscale
img = cv2.imread('ningning.jpeg', 0)

# Terapkan Sobel filter untuk mendeteksi tepi secara horizontal dan vertikal
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)  # Sobel X (Deteksi tepi horizontal)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)  # Sobel Y (Deteksi tepi vertikal)

# Konversi hasil deteksi tepi ke bentuk absolut dan tipe 8-bit (0-255) agar mudah ditampilkan
sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.convertScaleAbs(sobely)

# Hitung histogram untuk Sobel X dan Sobel Y
hist_sobelx = cv2.calcHist([sobelx], [0], None, [256], [0, 256])
hist_sobely = cv2.calcHist([sobely], [0], None, [256], [0, 256])

# Tampilkan hasil deteksi tepi menggunakan Matplotlib dan histogramnya
plt.figure(figsize=(12, 8))

# Sobel X (Deteksi Tepi Horizontal)
plt.subplot(2, 2, 1)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X (Deteksi Tepi Horizontal)')
plt.axis('off')

# Histogram Sobel X
plt.subplot(2, 2, 2)
plt.plot(hist_sobelx, color='black')
plt.title('Histogram Sobel X')
plt.xlim([0, 256])

# Sobel Y (Deteksi Tepi Vertikal)
plt.subplot(2, 2, 3)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y (Deteksi Tepi Vertikal)')
plt.axis('off')

# Histogram Sobel Y
plt.subplot(2, 2, 4)
plt.plot(hist_sobely, color='black')
plt.title('Histogram Sobel Y')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

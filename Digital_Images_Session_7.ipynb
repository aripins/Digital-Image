```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Memuat gambar
image = cv2.imread('wonyoung.jpg', cv2.IMREAD_GRAYSCALE)

# Menampilkan gambar asli
plt.figure(figsize=(6, 4))
plt.imshow(image, cmap='gray')
plt.title("Gambar Asli")
plt.axis('off')
plt.show()

# Menampilkan histogram citra asli
plt.figure(figsize=(6, 4))
plt.hist(image.ravel(), bins=256, range=[0, 256])
plt.title("Histogram Citra Asli")
plt.show()
```

```python
# Melakukan histogram equalization pada citra
equalized_image = cv2.equalizeHist(image)

# Menampilkan gambar hasil equalization
plt.figure(figsize=(6, 4))
plt.imshow(equalized_image, cmap='gray')
plt.title("Gambar Hasil Equalization")
plt.axis('off')
plt.show()
```

```python
# Menampilkan histogram dari citra hasil equalization
plt.figure(figsize=(6, 4))
plt.hist(equalized_image.ravel(), bins=256, range=[0, 256])
plt.title("Histogram Hasil Equalization")
plt.show()
```

```python
# Menampilkan perbandingan citra asli dan citra hasil equalization
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Gambar Asli")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title("Gambar Hasil Equalization")
plt.axis('off')

plt.show()
```
